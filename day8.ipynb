```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("transactions.csv")
df
```

```python
#1 total sales amount across all trasactions
total_sales = df["t_amt"].sum()
print(total_sales)
```

```python
#2 Convert t_date column to datetime (MM-DD-YYYY format)
df["t_date"] = pd.to_datetime(df["t_date"], format="%m-%d-%Y", errors="coerce")

# Drop rows where date conversion failed (if any)
df = df.dropna(subset=["t_date"])

# Extract Year-Month
df["month"] = df["t_date"].dt.to_period("M")

# Group by month and sum transaction amount
monthly_sales = df.groupby("month")["t_amt"].sum()

# Find month with highest total sales
highest_month = monthly_sales.idxmax()
highest_amount = monthly_sales.max()

print("Month with Highest Total Transaction Amount:", highest_month)
print("Total Sales Amount:", round(highest_amount, 2))
```

```python
#3 Calculate average transaction amount per customer
avg_per_customer = df.groupby("cust_id")["t_amt"].mean()
print("Average transaction amount per customer:")
print(avg_per_customer.head())
overall_avg = avg_per_customer.mean()
print("\nOverall Average Transaction Amount per Customer:", round(overall_avg, 2))
```

```python
#28.Is there a particular season when sports equipment sales spike?
# Converting date to datetime
df['t_date'] = pd.to_datetime(df['t_date'])

# Extracting month
df['month'] = df['t_date'].dt.month

# Filtering required services
sports_df = df[df['services'].isin([
    'Team Sports',
    'Outdoor Recreation',
    'Exercise & Fitness'
])]

# Monthly revenue for these services
monthly_revenue = sports_df.groupby(['services', 'month'])['t_amt'].sum()

# Sort to find highest revenue months
monthly_revenue.sort_values(ascending=False).reset_index(name="monthly_revenue")
```

```python
#33.Should more inventory be stocked for Exercise & Fitness products given their popularity?

ex_df = df[df['services'] == 'Exercise & Fitness']

# 1. Transaction Count
transaction_count = ex_df.shape[0]

# 2. Total Revenue
total_revenue = ex_df['t_amt'].sum()

# 3. Monthly Growth Trend
monthly_trend = ex_df.groupby('month')['t_amt'].sum().sort_index()

print("Transaction Count:", transaction_count)
print("Total Revenue:", total_revenue)
print("\nMonthly Sales Trend:")
print(monthly_trend)
```

```python
#35.Are there underperforming service categories that need promotional offers?

# Ensure amount is numeric
df['t_amt'] = pd.to_numeric(df['t_amt'], errors='coerce')

# Total revenue per service
service_revenue = df.groupby('services')['t_amt'].sum()

# Sort ascending to find lowest revenue categories
underperforming = service_revenue.sort_values().head(5).reset_index(name="underperforming")

print("Underperforming Services (Lowest Revenue):")
print(underperforming)
```
